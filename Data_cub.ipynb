```python
import pandas as pd
import numpy as np
import json
import os
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')

print("[SYSTEM] Environment ready. All libraries imported.")
```

```python
data = {
    "Product": ["TV","Laptop","TV","Speaker","Tablet","Laptop","TV","Speaker","Tablet","TV"],
    "City": ["Kathmandu","Kathmandu","Pokhara","Biratnagar","Kathmandu","Pokhara","Kathmandu","Biratnagar","Kathmandu","Pokhara"],
    "Store_Address": [
        "New Road 12",
        "Durbar Marg 5",
        "Lakeside 21",
        "Main Road 8",
        "Putalisadak 12",
        "Chipledhunga 33",
        "Thamel 5",
        "Biratnagar Bazaar 8",
        "Baneshwor 12",
        "Mahendrapool 21"
    ],
    "Season": ["Q1","Q2","Q1","Q3","Q1","Q4","Q1","Q3","Q2","Q1"],
    "Sales": [12000,15000,11000,5000,8000,14000,13000,6000,9000,12500],
    "Quantity": [12,15,11,5,8,14,13,6,9,12]
}

df = pd.DataFrame(data)
df
```

```python
# City-level aggregation
city_level = df.groupby("City")["Sales"].sum()
city_level

# Drill down to Store-level
store_level = df.groupby(["City","Store_Address"])["Sales"].sum()
store_level
```

```python
# Seasonal totals
season_sales = df.groupby("Season")["Sales"].sum()
season_sales

# Yearly total (sum of all seasons)
yearly_sales = df["Sales"].sum()
yearly_sales
```
